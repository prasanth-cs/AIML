```python
# CODE FRO REMOVING MISSING DATA FROM THE DATA
import pandas as pd

#EXAMPLE DATAFRAME

data=pd.DataFrame({'A':[1,2,3,None,4],'B':[6,7,8,None,0],'C':[11,23,45,None,53]})
data
```

```python
#REMOVE ROWS WITH MISSING VALUES
data_clean = data.dropna()
print(data_clean)
```

```python
#REPLACING MISSING WITH COLUMN MEAN
data_filled = data.fillna(data.mean())
print(data_filled)
```

```python
#NORMALIZATION
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data)
print(data_normalized)
```

```python
#LABEL ENCODER
from sklearn.preprocessing import LabelEncoder
data=pd.DataFrame({'Color':['red','blue','green','red',]})
encoder = LabelEncoder()
data['Color_Label'] = encoder.fit_transform(data['Color'])
print(data)
```

```python
#ONE HOT ENCODING
data=pd.DataFrame({'Color':['red','blue','green','red',]})
data_onehot = pd.get_dummies(data, columns=['Color'])
print(data_onehot)
```

```python
#OUTLIER DETECTION
import numpy as np
from scipy import stats
data = [1,2,2,2,2,3,1,1,15,2,2,2,3,1,22,1,2]
mean=np.mean(data)
std=np.std(data)
print('mean of data set is ',mean)
print('Std deviation is',std)
```

```python
#CALCULATE Z SCORE . FI Z SCORE>3, PRINT IT AS AN OUTLIERS.
#CALCULATE Z SCORE . FI Z SCORE>3, PRINT IT AS AN OUTLIERS.
threshold = 3
outlier=[]
for i in data:
  z=(i-mean)/std
  if z> threshold:
    outlier.append(i) #Corrected the indentation here to align with the 'if' block
print('outliers are',outlier)
```

```python
#TEXT DATA PROCESSING
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
corpus = ["This is a sentence.","This is another sentence."]
filtered_corpus = [''.join([word for word in text.split() if word not in ENGLISH_STOP_WORDS]) for text in corpus]
print(filtered_corpus)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
corpus = ["running", "ranner", "run"]
stemmed_words = [ps.stem(word) for word in corpus]
print(stemmed_words)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'data' is your DataFrame
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')

sns.countplot(ax=axes[0, 0], x='Fuel_Type', data=data, color='blue',
              order=data['Fuel_Type'].value_counts().index)
sns.countplot(ax=axes[0, 1], x='Transmission', data=data, color='blue',
              order=data['Transmission'].value_counts().index)
sns.countplot(ax=axes[1, 0], x='Owner_Type', data=data, color='blue',
              order=data['Owner_Type'].value_counts().index)
sns.countplot(ax=axes[1, 1], x='Location', data=data, color='blue',
              order=data['Location'].value_counts().index)
sns.countplot(ax=axes[2, 0], x='Brand', data=data, color='blue',
              order=data['Brand'].value_counts().head(20).index)
sns.countplot(ax=axes[2, 1], x='Model', data=data, color='blue',
              order=data['Model'].value_counts().head(28).index)

# Rotate labels for better readability
for ax in axes.flatten():
    ax.tick_params(labelrotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to make room for the title
plt.show()
```
